```python
import os
import sys
from download_videos import download_videos
from folder_select import select_folder_dialog

def main():
    # Ensure a directory argument is provided
    if len(sys.argv) < 2:
        print(f"Usage: {sys.argv[0]} <directory>")
        sys.exit(1)

    # Directory containing the text files (train.txt, val.txt, test.txt)
    DIR = select_folder_dialog()

    # Download videos for train.txt
    print("Downloading Train set")
    download_videos(os.path.join(DIR, 'train.txt'), os.path.join(DIR, 'videos/train'))

    # Download videos for val.txt
    print("Downloading Val set")
    download_videos(os.path.join(DIR, 'val.txt'), os.path.join(DIR, 'videos/val'))

    # Download videos for test.txt
    print("Downloading Test set")
    download_videos(os.path.join(DIR, 'test.txt'), os.path.join(DIR, 'videos/test'))

    # Additional part to mimic the last part of the original script
    dst_dir = os.path.join(DIR, 'intervals')
    os.makedirs(dst_dir, exist_ok=True)

if __name__ == "__main__":
    main()
```
